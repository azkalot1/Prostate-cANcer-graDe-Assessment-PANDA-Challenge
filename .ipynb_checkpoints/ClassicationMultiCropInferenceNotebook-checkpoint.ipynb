```python
name = 'afeae451e576086d10c45a41545f7212'
img = openslide.OpenSlide(os.path.join(TRAIN, name+'.tiff'))
mask = openslide.OpenSlide(os.path.join(MASKS, name+'_mask.tiff'))
img = img.read_region(
    (0, 0),
    img.level_count - 1,
    img.level_dimensions[-1])
mask = mask.read_region(
    (0, 0),
    mask.level_count - 1,
    mask.level_dimensions[-1])
```

```python
img = np.asarray(img)[..., :3]
mask = np.asarray(mask)[..., :3]
tiles = tile(img, mask=mask, N=24)
tiled_images = (1.0 - tiles['img']/255.0)
tiled_images = (tiled_images - mean)/std
```

```python
class Model(nn.Module):    
    def __init__(self, arch='resnet34', n_classes=6, **kwargs):
        super().__init__()
        m = timm.create_model(
            arch, 
            **kwargs)
        self.enc = nn.Sequential(*list(m.children())[:-2])       
        nc = list(m.children())[-1].in_features
        self.head = nn.Sequential(
            nn.AdaptiveMaxPool2d((1,1)), 
            nn.ReLU(),
            nn.Flatten(), 
            nn.Dropout(0.5),
            nn.Linear(nc, n_classes))
        
    def forward(self, x):
        print('Initial,  ', x.size())
        shape = x[0].shape
        n = len(x[0])
        x = x.view(-1, shape[1], shape[2], shape[3])
        #x: bs*N x 3 x 128 x 128
        x = self.enc(x)
        # should be: bs*N x C x 4 x 4
        shape = x.shape
        #concatenate the output for tiles into a single map
        x = x.view(-1, n, shape[1], shape[2], shape[3])
        x = x.permute(0, 2, 1, 3, 4).contiguous()
        x = x.view(-1, shape[1], shape[2]*n, shape[3])
        # should be: bs x C x N*4 x 4
        x = self.head(x)
        print(x.size())
        return x
```

```python
import albumentations as A
import json
```

```python
with open('./configs/light_configs_noNorm.json') as f:
    augs = A.from_dict(json.load(f))
```

```python
def create_transformer(transformations, N):
    target = {}
    for i in range(1, N):
        target['image' + str(i)] = 'image'
    return A.Compose(transformations, p=1, additional_targets=target)
```

```python
augs = create_transformer(augs, 24)
target_names = ['image' + str(i) if i > 0 else 'image'
                for i in range(len(tiled_images))]
```

```python
tiled_images = dict(
    zip(
        target_names,
        tiled_images)
)
```
